# F) Clean the dataset

F1) convert the necessary columns to a 'NEW' columns from existing dataframe

```python
#date will be on yyyy-mm-dd 
select_df['new_uen_issue_date'] = pd.to_datetime(create_df.uen_issue_date)
select_df['new_registration_date'] = pd.to_datetime(create_df.registration_incorporation_date)
select_df.head()
```

F2)create a 'new year month', 'new month' and 'new year' columns from registration_date

```python
select_df['new_year'] = pd.DatetimeIndex(select_df['new_registration_date']).year
select_df['new_month'] = pd.DatetimeIndex(select_df['new_registration_date']).month
select_df['new_year_month'] = pd.DatetimeIndex(select_df['new_registration_date']).to_period('M')
select_df.head()
```

```python
new_set = select_df.dropna(axis=1,how='all')
new_set.head()
```

```python
new_set.info()
```

F3) name a variable and convert datetime from registration date columns

```python
registration_date = pd.to_datetime(new_set['new_registration_date'])
registration_date
```

F4) Create a new column and name as 'no_of_years' to calculate the years of the company since the start of registration

```python
from datetime import datetime
current = datetime.now()
current
```

```python
new_set['no_of_years'] = current - registration_date
new_set['no_of_years'] = new_set['no_of_years'].dt.days #turn timedelta into int number of days
new_set['no_of_years'] = round(new_set['no_of_years']/365.25) #approximation with days - 1 year = 365.25 days
new_set.head()
```

f5) replace 'NaN' to empty fill

```python
new_set = new_set.replace(np.nan, '')
new_set.head()
```

F6) Name a new variable as 'new_df'

```python
new_df = new_set
new_df.head()
```
